# **Rotten Tomatoes: Non Personalized Recommender**

Now some of the non-personalized recommenders are:
1. **Rotten Tomatoes**
2. **Metacritic**
3. **IMDB**

Rotten Tomatoes – Aggregates film and TV reviews from critics and audiences, scoring content based on the percentage of positive reviews ("Tomatometer").

Metacritic – Combines critic reviews into a weighted average score (0–100) for movies, games, and TV, offering a "Metascore" based on review quality and prominence.

IMDb – A massive user-driven database for movies, TV shows, and celebrities, featuring ratings, reviews, and industry information, with scores based on global user votes.


### **Technique Used:**
Rotten Tomatoes aggregates **critic reviews** and generates two scores:
- **Tomatometer**: % of approved critics who gave a positive review.
- **Audience Score**: % of regular users who rated positively.

This is a **non-personalized popularity-based recommender**. It doesn't rely on user profile history but instead displays what's highly rated.

### **My Preferred Choice:**

# **Attacks on Recommender Systems**

### **Article Summary:**

In 2017, the film The Promise, which depicts the Armenian Genocide during the fall of the Ottoman Empire, became the center of a politically motivated campaign on IMDb. Before the film had even been released to the general public, tens of thousands of users flooded IMDb with 1-star ratings, while an opposing group countered with 10-star ratings. The sudden spike in extreme ratings was not based on viewership or artistic merit but driven largely by political agendas surrounding the film’s sensitive historical content.

This manipulation raised questions about the vulnerability of IMDb's open rating system, which allows users to rate films even if they haven’t been verified as viewers, unlike platforms like Rotten Tomatoes which may verify tickets for audience scores.

### **Similar Example:**

In early 2019, Rotten Tomatoes became the target of a widespread review bombing campaign against Captain Marvel, a Marvel Studios film starring Brie Larson. Before the film’s release, users flooded the platform with negative audience comments and low “Want to See” scores, despite not having seen the movie. Many of these comments were driven not by the film’s content, but by backlash to Larson’s public statements on diversity and inclusion in film criticism.

Unlike critic reviews, which are verified and only posted after the movie is released, audience scores and “Want to See” metrics were open to anyone—no ticket or viewing confirmation required. This made it easy for coordinated groups to manipulate the film’s public image before its release.

**In response, Rotten Tomatoes took unprecedented action:**

- It disabled public commenting and audience scores prior to release.

- It adjusted the “Want to See” metric to avoid showing it as a rating.

- Later, it began verifying tickets for audience ratings (via platforms like Fandango).

These changes were designed to increase authenticity and reduce politically or socially motivated manipulation that had nothing to do with the actual quality of the film.

### **How to Prevent Abuse:**
To mitigate such attacks:
1. **Delay user ratings until after official release**
2. **Weight ratings based on verified purchase/viewership**
3. **Use anomaly detection** (e.g., sudden rating spikes)
4. **Rate-limiting for reviews per IP/user**
5. **Trust-based user reputation systems**

These strategies can improve the resilience of recommenders against coordinated attacks.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Simulated data
days = np.arange(-7, 8)  # 15 days: -7 to +7

# Create exactly 15 ratings: 5 before, 5 on/around release, 5 after
attack_ratings = np.random.normal(2, 0.3, 5)          # Before release
release_ratings = np.random.normal(6.5, 0.6, 5)       # On release
post_release_ratings = np.random.normal(7, 0.4, 5)    # After release

ratings = np.concatenate([attack_ratings, release_ratings, post_release_ratings])

# Create DataFrame
df = pd.DataFrame({'Day': days, 'Rating': ratings})

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Day', y='Rating', marker='o')
plt.axvline(0, color='red', linestyle='--', label='Release Day')
plt.title("IMDb Ratings Trend: Pre-Release Attack Example")
plt.xlabel("Days Relative to Release")
plt.ylabel("User Rating")
plt.legend()
plt.grid(True)
plt.show()
```
